# Capture camera image and convert to base64 string

```python
import cv2
import base64

# Capture an image from the camera
camera = cv2.VideoCapture(0)

# Check if the camera opened successfully
if not camera.isOpened():
    print("Could not open camera")
    exit()

while True:
    # Capture a single frame
    ret, frame = camera.read()

    # Check if the frame was captured successfully
    if not ret:
        print("Failed to grab frame")
        break

    # Display the live camera feed for preview
    cv2.imshow("Press 'c' to capture or 'q' to quit", frame)

    # Wait for a key event 
    key = cv2.waitKey(1)

    # If 'c' is pressed, capture the image
    if key == ord('c'):
        captured_frame = frame.copy()
        break
    # If 'q' is pressed, quit the application
    elif key == ord('q'):
        print("User quit before capturing image.")
        camera.release()
        cv2.destroyAllWindows()
        exit()

# Release the camera
camera.release()

# Display the captured frame
cv2.imshow("Captured Image", captured_frame)

# Save the image to a file
cv2.imwrite("captured_image.jpg", captured_frame)

# Wait for a key event and then close the image window
cv2.waitKey(0)
cv2.destroyAllWindows()

# Convert the image to a base64 string
with open("captured_image.jpg", "rb") as image_file:
    base64_string = base64.b64encode(image_file.read()).decode()

# Print the base64 string
print(base64_string)
```

# For better chance of detecting TICKER, i'm combining results from all azure services above 

```python
from azure.cognitiveservices.vision.computervision import ComputerVisionClient
from msrest.authentication import CognitiveServicesCredentials
import json

# Azure Computer Vision setup
endpoint = 'https://wimhackathon2023.cognitiveservices.azure.com/'
key = 'bbcbbed83121438a92eaaf6f3bb4caa6'
computervision_client = ComputerVisionClient(endpoint, CognitiveServicesCredentials(key))


# URL of the image to be analyzed
image_url = 'https://media1.popsugar-assets.com/files/thumbor/XzWrVYmG-p3GZQDWBMlCuJPBwP4/fit-in/1024x1024/filters:format_auto-!!-:strip_icc-!!-/2019/06/18/865/n/1922441/4b3090b55d093fa651ef63.08756026_71HlJVxFl7L._SL1500_/i/Kirkland-Signature-Premium-Drinking-Water.jpg'

# Analyze image for brands, description, and tags
image_features = ["brands", "description", "tags"]
analysis_result = computervision_client.analyze_image(image_url, visual_features=image_features)

# Extract printed text with OCR
ocr_result = computervision_client.recognize_printed_text(url=image_url)

# Organize results into a dictionary
result_data = {
    'brands': [brand.name for brand in analysis_result.brands] if hasattr(analysis_result, 'brands') and analysis_result.brands else [],
    'description': analysis_result.description.captions[0].text if hasattr(analysis_result, 'description') and analysis_result.description.captions else "",
    'tags': [tag.name for tag in analysis_result.tags] if hasattr(analysis_result, 'tags') and analysis_result.tags else [],
    'extracted_text': [
        {
            "words": [{"text": word.text} for line in region.lines for word in line.words] 
        } 
        for region in ocr_result.regions
    ]
}

# Convert the results to JSON
result_json = json.dumps(result_data, indent=2)

# Print the JSON result
print("Result JSON:")
print(result_json)


```
